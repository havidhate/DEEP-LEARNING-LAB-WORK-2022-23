```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate input data
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = sigmoid(X)

# Plot in 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Sigmoid(X)')
ax.set_title('3D Plot of Sigmoid Function')
plt.show()

# Plot in 2D
plt.figure(figsize=(8, 6))
plt.plot(x, sigmoid(x))
plt.xlabel('X')
plt.ylabel('Sigmoid(X)')
plt.title('2D Plot of Sigmoid Function')
plt.grid(True)
plt.show()
```
