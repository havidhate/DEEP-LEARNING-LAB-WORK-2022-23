```python
import numpy as np
import matplotlib.pyplot as plt

# Define the feedforward network function
def feedforward_network(x, y):
    return x**2 + y**2

# Generate input data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute the output of the feedforward network for each (x, y) pair
Z = feedforward_network(X, Y)

# Create a contour plot
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(contour)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour Plot of FeedForward Network')
plt.show()
```
